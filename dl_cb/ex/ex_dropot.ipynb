```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../csv/sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.isna().sum()
```

```python
X = df.drop(60, axis='columns')
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first=True)
y = y.astype(np.uint8)
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X_train.shape, X_test.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(60, input_dim=60, activation='relu'),
    tf.keras.layers.Dense(30, input_dim=60, activation='relu'),
    tf.keras.layers.Dense(15, input_dim=60, activation='relu'),
    tf.keras.layers.Dense(1, input_dim=60, activation='sigmoid')]
)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=120)

model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
print(classification_report(y_test, y_pred))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(60, input_dim=60, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(30, input_dim=60, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(15, input_dim=60, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, input_dim=60, activation='sigmoid')]
)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=120)

model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
print(classification_report(y_test, y_pred))
```
